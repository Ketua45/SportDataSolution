# Création des fichiers de tables SQL

```python
# Table département

data_departement = pd.DataFrame(data_rh_sport["BU"].unique(), columns=["nom_departement"])

data_departement = data_departement.reset_index().rename(columns={"index": "departement_id"})

data_departement.to_csv('/home/romain/formation_data_engineer/Projet_12_SportDataSolution/data/tables_sql/departements.csv', index=False)

data_departement
```

```python
# Table contrat

data_contrat = pd.DataFrame(data_rh_sport["Type de contrat"].unique(), columns=["nom_contrat"])

data_contrat = data_contrat.reset_index().rename(columns={"index": "contrat_id"})

data_contrat.to_csv('/home/romain/formation_data_engineer/Projet_12_SportDataSolution/data/tables_sql/contrats.csv', index=False)

data_contrat
```

```python
# Table moyen de déplacement

data_moyen_deplacement = pd.DataFrame(data_rh_sport["Moyen de déplacement"].unique(), columns=["nom_moyen_deplacement"])

data_moyen_deplacement = data_moyen_deplacement.reset_index().rename(columns={"index": "moyen_deplacement_id"})

data_moyen_deplacement.to_csv('/home/romain/formation_data_engineer/Projet_12_SportDataSolution/data/tables_sql/moyens_deplacement.csv', index=False)

data_moyen_deplacement
```

```python
# Création table employés avec les clés étrangères

data_employes = data_rh_sport[["ID salarié", "Nom", "Prénom", "Date de naissance", "Date d'embauche", "Type de contrat", "BU", "Moyen de déplacement", "Nombre de jours de CP", "Salaire brut", "Adresse du domicile"]]

data_employes = (
    data_employes
    .join(data_contrat.set_index("nom_contrat"), on="Type de contrat", how="right")
    .join(data_departement.set_index("nom_departement"), on="BU", how="right")
    .join(data_moyen_deplacement.set_index("nom_moyen_deplacement"), on="Moyen de déplacement", how="right")
    .drop(columns=["Type de contrat", "BU", "Moyen de déplacement"])
)


# Changement des noms de colonnes pour correspondre à la convention snake_case et à la bdd

data_employes = data_employes.rename(columns={
    "ID salarié": "employe_id",
    "Nom": "nom_employe",
    "Prénom": "prenom_employe",
    "Date de naissance": "date_naissance_employe",
    "BU": "departement_id",
    "Date d'embauche": "date_embauche",
    "Salaire brut": "salaire",
    "contrat_id": "type_contrat_id",
    "Adresse du domicile": "adresse",
    "Moyen de déplacement": "moyen_deplacement_id",
    "Nombre de jours de CP": "nombre_conges"
})
    
data_employes.to_csv('/home/romain/formation_data_engineer/Projet_12_SportDataSolution/data/tables_sql/employes.csv', index=False)

data_employes
```

# Création de données STRAVA Like

```python
# Fonction de création de données virtuelles

def StravaLikeDataGen(dataset_employes, nombre_activites):

    """""""""""""""""""""
    Fonction qui permet de générer des activités fictives sur le modèle de Strava, pour les salariés de l'entreprise.
     - dataset_employes : dataframe contenant les données des employés (doit contenir une colonne "employe_id" pour faire le lien avec les activités)
     - nombre_activites : nombre d'activités fictives à générer
    """""""""""""""""""""
    import random
    import pandas as pd

    sports = ["Course à pied", "Vélo", "Natation", "Marche", "Trotinette", "Voile"]
    id_employes = dataset_employes["employe_id"]
    distance = range(1,15)
    date = pd.date_range(start='2025-01-01', end='2025-12-31', freq='D')
    temps_exercice = range(600,3600)

    commentaires = [
        "Super séance, je me suis senti(e) en pleine forme !",
        "Pas mal, mais j'ai eu du mal à trouver mon rythme.",
        "A refaire très vite !",
        "J'ai adoré cette activité, c'était vraiment génial !"]
    
    print("Tentative d'insertion de", nombre_activites, "activités...")
    for _ in range(nombre_activites):
        activite = pd.DataFrame({
            "employe_id": [random.choice(id_employes)],
            "type_pratique_sportive": [random.choice(sports)],
            "distance": [random.choice(distance)],
            "date": [random.choice(date)],
            "temps_ecoule": [random.choice(temps_exercice)],
            "commentaire": [random.choice(commentaires)]
        })

        # Récupérer le dernier id présent dans la base
        last_id_query = "SELECT COALESCE(MAX(id_evenement_sportif), 0) FROM pratique_sport"
        last_id = pd.read_sql(last_id_query, con=engine).iloc[0, 0]

        activite.index = [last_id + 1]
        
        activite.to_sql(
            name="pratique_sport",
            con=engine,
            if_exists="append",
            index=True, 
            index_label = "id_evenement_sportif"
        )
    
    print(nombre_activites, "activités insérées avec succès")
    
    pass
```
